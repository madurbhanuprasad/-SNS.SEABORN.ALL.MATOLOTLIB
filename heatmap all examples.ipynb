```python
import seaborn as sns
import matplotlib.pyplot as plt

# Basic heatmap using a random correlation matrix
corr_matrix = sns.load_dataset("iris").corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Basic Heatmap of Correlation Matrix")
plt.show()

```

```python
#Certainly! The .corr() method calculates the correlation matrix for a given DataFrame.
#Correlation measures the strength and direction of the linear relationship between pairs 
#of numerical variables. Let's go through a basic example:



import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load a sample dataset
data = sns.load_dataset("iris")

# Calculate the correlation matrix
corr_matrix = data.corr()

# Display the correlation matrix
print("Correlation Matrix:")
print(corr_matrix)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")#"viridis""plasma""inferno""magma""cividis"
plt.title("Heatmap of Correlation Matrix")
plt.show()
```

```python
# Customized heatmap with color palette and annotations
flights = sns.load_dataset("flights").pivot_table(index="month", columns="year", values="passengers")
sns.heatmap(flights, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Customized Heatmap of Flight Passengers")
plt.show()
```

```python
#The fmt parameter in Seaborn's heatmap function (sns.heatmap) is used to format the annotations that are displayed on the heatmap cells. It allows you to control the format of the numbers that appear in the annotations.

In the context of fmt="d", the "d" stands for "decimal." 
This format specifier is commonly used in Python's string formatting to 
indicate that a number should be displayed as an integer (decimal) without
any decimal places or additional formatting.

For example, if you have a heatmap that displays numerical values and you set
fmt="d", the numbers in the annotations will be formatted as integers without
decimal places. Here's an example
```

```python
#Hierarchical Clustering Heatmap:
# Hierarchical clustering heatmap using the seaborn's clustering functions
flights = sns.load_dataset("flights").pivot_table(index="month", columns="year", values="passengers")
sns.clustermap(flights, cmap="viridis")
plt.title("Hierarchical Clustering Heatmap of Flight Passengers")
plt.show()
```

```python
# Annotated heatmap with custom color palette and annotations
data = sns.load_dataset("titanic")
pivot_table = data.pivot_table(index="class", columns="sex", values="age")
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="coolwarm")
plt.title("Annotated Heatmap of Age by Class and Sex")
plt.show()
```

```python
stock_data = {
    "2021-01-01": {"AAPL": 150, "GOOGL": 2500, "AMZN": 3200},
    "2021-01-02": {"AAPL": 155, "GOOGL": 2520, "AMZN": 3250},
    
}
```

```python
#Convert the dictionary to a DataFrame:
data = pd.DataFrame.from_dict(stock_data, orient="index")
#Reset the index and rename the index column to "Date":
data.reset_index(inplace=True)
data.rename(columns={"index": "Date"}, inplace=True)
data["Date"] = pd.to_datetime(data["Date"])
#Melt the DataFrame to reshape it for the pivot:
melted_data = data.melt(id_vars=["Date"], var_name="StockSymbol", value_name="StockPrice")
#Pivot the melted data to create a pivot table:
pivot_table = melted_data.pivot_table(index="Date", columns="StockSymbol", values="StockPrice")
#Create the heatmap using Seaborn:
plt.figure(figsize=(12, 6))
sns.heatmap(pivot_table, cmap="YlGnBu")
plt.title("Time Series Heatmap of Stock Prices")
plt.xlabel("Stock Symbol")
plt.ylabel("Date")
plt.show()
```
